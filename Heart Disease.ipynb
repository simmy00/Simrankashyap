## dataset details 
- age
- sex
- Chest pain type (4 values)
    - value 0:Typical angina
    - value 1:atypical angina
    - value 2:non-anginal pain
    - value 3:asymptomatic
- trestbps:resting blood pressure
- chol: serum cholestoral in mg/dl
- fbs: fasting blood sugar > 120 mg/dl ( value = 0,1)
- restecg: resting electrocardiographic results (values 0,1,2) 
    - value 0 :normal
    - value 1 :having ST-T wave Abnormality 
    - value 2 :showing probable or definite left ventricular hypertrophy by Etes's criteria 
- thalach :maximum heart rate achieved 
- exang : exercise induced angina
- oldpeak: oldpeak = ST depression induced by exercise relative to rest
- slope : the slope of the peak exercise ST segment
    - value 0 : upsloping
    - value 1 : Flat
    - value 3 : Downsloping
- ca : number of major vessels (0-3) colored by flourosopy
- thal: thal: 0 = normal; 1 = fixed defect; 2 = reversable defect 
- target: 0 = Less chances of heart attack , 1 = more chances of heart attack

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert 'sex' column to string type
df['sex'] = df['sex'].astype(str)

# Grouping the data by 'target' and 'sex' and counting the occurrences
gender_distribution = df.groupby(['target', 'sex']).size().reset_index(name='count')

# Plotting the distribution
plt.figure(figsize=(8, 6))
sns.barplot(data=gender_distribution, x='target', y='count', hue='sex')
plt.title('Gender Distribution According to Target Variable')
plt.xlabel('Target')
plt.ylabel('Count')
plt.xticks([1,0],['Female','Male'])
plt.legend(title='Sex', loc='upper right')
plt.show()
```

# Distribution of Resting blood pressure according to target 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert 'cp' and 'target' columns to string type
df['trestbps'] = df['trestbps'].astype(str)
df['target'] = df['target'].astype(str)

# Grouping the data by 'cp' and 'target' and calculating the count
grouped_df = df.groupby(['trestbps', 'target']).size().reset_index(name='count')

# Plotting the distribution
plt.figure(figsize=(20, 6))
sns.barplot(data=grouped_df, x='trestbps', y='count', hue='target')
plt.title('Distribution of Resting blood pressure by Target Variable')
plt.xlabel('Resting blood pressure (trestbps)')
plt.ylabel('Count')
plt.legend(title='Target', loc='upper right')

plt.show()
```

```python
# Value counts for categorical variables
print("\nValue counts for categorical variables:")
for col in ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal', 'target']:
    print("\n", col)
    print(df[col].value_counts())
```

```python
# Univariate analysis: Histograms for numerical variables
numerical_vars = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
for col in numerical_vars:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df, x=col, bins=20, kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Univariate analysis: Bar plots for categorical variables
categorical_vars = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal', 'target']
for col in categorical_vars:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=col)
    plt.title(f'Countplot of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
# Convert 'target' variable to numeric if needed
df['target'] = pd.to_numeric(df['target'], errors='coerce')

# Convert 'col' variable to numeric if needed
for col in numerical_vars:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Bivariate analysis: Box plots for numerical variables vs. target variable
for col in numerical_vars:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, x='target', y=col)
    plt.title(f'Boxplot of {col} by Target')
    plt.xlabel('Target')
    plt.ylabel(col)
    plt.show()
```

```python
# Bivariate analysis: Stacked bar plots for categorical variables vs. target variable
for col in categorical_vars:
    if col != 'target' and col != 'oldpeak':  # Exclude 'oldpeak' from countplot visualization
        plt.figure(figsize=(8, 6))
        sns.countplot(data=df, x='target', hue=col)
        plt.title(f'Countplot of {col} by Target')
        plt.xlabel('Target')
        plt.ylabel('Count')
        plt.legend(title=col, loc='upper right')
        plt.show()
```
